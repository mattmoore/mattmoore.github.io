```python
import librosa, librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd

plt.style.use('dark_background')

signal, sr = librosa.load("Haunting_song_of_humpback_whales-youtube-W5Trznre92c.wav")

ipd.Audio(signal, rate=sr)
```

```python
# this is the number of samples in a window per fft
n_fft = 2048
# The amount of samples we are shifting after each fft
hop_length = 512
# Short-time Fourier Transformation on our audio data
audio_stft = librosa.core.stft(signal, hop_length=hop_length, n_fft=n_fft)
spectrogram = np.abs(audio_stft)
# Here we're converting raw amplitude to decibels to make it easier to see the spectrogram
power_to_db = librosa.power_to_db(spectrogram, ref=np.max)

plt.figure(figsize=(20, 5))
# Insted of passing the spectrogram with raw power to the specshow function, here we pass the decibels
librosa.display.specshow(power_to_db, sr=sr, x_axis="time", y_axis="hz", hop_length=hop_length)
plt.colorbar(label="dB")
plt.title("Spectrogram (dB)", fontdict=dict(size=18))
plt.xlabel("Time", fontdict=dict(size=15))
plt.ylabel("Frequency", fontdict=dict(size=15))
plt.show()
```
